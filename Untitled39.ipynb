```python
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import random
from collections import Counter
import numpy as np


class_mapping = {
    0: 'Housekeeping Gene',
    1: 'Oncogene',
    2: 'Tumor Suppressor',
    3: 'Immune Response Gene',
    4: 'Metabolic Gene',
    5: 'Signal Transduction Gene',
    6: 'Unknown/Other'
}

file_path = "/content/human_data.txt"
data = []

def calculate_gc_content(seq):
    gc = seq.count("G") + seq.count("C")
    return gc / len(seq) if len(seq) > 0 else 0

with open(file_path, 'r') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        parts = line.split('\t') if '\t' in line else line.split(',')
        if len(parts) == 2:
            seq, label = parts[0].strip(), parts[1].strip()
            if all(base in "ATGC" for base in seq) and label.isdigit():
                gc_content = calculate_gc_content(seq)
                data.append((seq, int(label), gc_content))

df = pd.DataFrame(data, columns=["sequence", "label", "gc"])
if df.empty:
    raise ValueError("üö® Your file is empty or not properly formatted!")


plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='label', y='gc')
plt.title("üß™ GC Content Distribution by Gene Class")
plt.xlabel("Gene Class")
plt.ylabel("GC Content Ratio")
plt.show()

df['length'] = df['sequence'].apply(len)
plt.figure(figsize=(8, 5))
sns.histplot(df['length'], bins=30, kde=True, color="skyblue")
plt.title("üìè Sequence Length Distribution")
plt.xlabel("Sequence Length")
plt.ylabel("Frequency")
plt.show()

class_counts = df['label'].value_counts().sort_index()
labels = [class_mapping.get(i, str(i)) for i in class_counts.index]
plt.figure(figsize=(7, 7))
plt.pie(class_counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("üß¨ Distribution of Gene Classes")
plt.axis('equal')
plt.show()

MAX_SEQ_LEN = 512

class DNATokenizer:
    def __init__(self):
        self.vocab = {"A": 0, "T": 1, "G": 2, "C": 3}
    def encode(self, seq):
        return [self.vocab.get(base, 0) for base in seq[:MAX_SEQ_LEN]]

tokenizer = DNATokenizer()

class DNADataset(Dataset):
    def __init__(self, sequences, labels, gcs):
        self.sequences = sequences
        self.labels = labels
        self.gcs = gcs
    def __len__(self):
        return len(self.sequences)
    def __getitem__(self, idx):
        x = torch.tensor(tokenizer.encode(self.sequences[idx]), dtype=torch.long)
        y = torch.tensor(self.labels[idx], dtype=torch.long)
        gc = torch.tensor([self.gcs[idx]], dtype=torch.float)
        return x, gc, y

def collate_fn(batch):
    sequences, gcs, labels = zip(*batch)
    lengths = torch.tensor([len(seq) for seq in sequences], dtype=torch.long)
    padded = pad_sequence(sequences, batch_first=True, padding_value=0)
    return padded, torch.stack(gcs), torch.tensor(labels), lengths

class Attention(nn.Module):
    def __init__(self, hidden_dim):
        super().__init__()
        self.attn = nn.Linear(hidden_dim * 2, 1)

    def forward(self, rnn_output, lengths):
        attn_scores = self.attn(rnn_output).squeeze(-1)
        mask = torch.arange(rnn_output.size(1), device=rnn_output.device).unsqueeze(0) >= lengths.unsqueeze(1)
        attn_scores.masked_fill_(mask, -1e9)
        attn_weights = F.softmax(attn_scores, dim=1)
        context = torch.sum(rnn_output * attn_weights.unsqueeze(-1), dim=1)
        return context, attn_weights

class DNAClassifier(nn.Module):
    def __init__(self, vocab_size=4, embed_dim=128, hidden_dim=256, num_classes=7):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True, bidirectional=True)
        self.attention = Attention(hidden_dim)
        self.gc_fc = nn.Linear(1, 32)
        self.classifier = nn.Linear(hidden_dim * 2 + 32, num_classes)

    def forward(self, x, gc, lengths):
        embedded = self.embedding(x)
        packed_out, _ = self.lstm(embedded)
        context, attn_weights = self.attention(packed_out, lengths)
        gc_feat = F.relu(self.gc_fc(gc))
        combined = torch.cat((context, gc_feat), dim=1)
        out = self.classifier(combined)
        return out, attn_weights

X_train, X_test, y_train, y_test, gc_train, gc_test = train_test_split(
    df['sequence'], df['label'], df['gc'], test_size=0.2, random_state=42
)

train_dataset = DNADataset(X_train.tolist(), y_train.tolist(), gc_train.tolist())
test_dataset = DNADataset(X_test.tolist(), y_test.tolist(), gc_test.tolist())
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_dataset, batch_size=8, collate_fn=collate_fn)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DNAClassifier(num_classes=len(class_mapping)).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

print("\nüöÄ Training Started...")
for epoch in range(10):
    model.train()
    total_loss = 0
    for x, gc, y, lengths in train_loader:
        x, gc, y, lengths = x.to(device), gc.to(device), y.to(device), lengths.to(device)
        optimizer.zero_grad()
        out, _ = model(x, gc, lengths)
        loss = criterion(out, y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"‚úÖ Epoch {epoch+1}, Loss: {total_loss / len(train_loader):.4f}")

model.eval()
preds, labels = [], []
confidences = []

with torch.no_grad():
    for x, gc, y, lengths in test_loader:
        x, gc, y, lengths = x.to(device), gc.to(device), y.to(device), lengths.to(device)
        out, _ = model(x, gc, lengths)
        prob = F.softmax(out, dim=1)
        conf, pred = torch.max(prob, dim=1)
        preds.extend(pred.cpu().numpy())
        confidences.extend(conf.cpu().numpy())
        labels.extend(y.cpu().numpy())

print("\nüìä Classification Report:")
print(classification_report(labels, preds, target_names=[class_mapping[i] for i in sorted(class_mapping)]))

sample_seq, sample_gc, _ = test_dataset[0]
sample_seq = sample_seq.unsqueeze(0).to(device)
sample_gc = sample_gc.unsqueeze(0).to(device)
sample_len = torch.tensor([sample_seq.size(1)]).to(device)

with torch.no_grad():
    _, attn_weights = model(sample_seq, sample_gc, sample_len)
attn_weights = attn_weights.squeeze().cpu().numpy()

tokens = list(X_test.tolist()[0][:len(attn_weights)])
plt.figure(figsize=(15, 2))
sns.heatmap([attn_weights], cmap='viridis', xticklabels=tokens, yticklabels=['Attention'])
plt.xticks(rotation=90)
plt.title("üß† Attention Map on DNA Sequence")
plt.show()

true_count = Counter(labels)
pred_count = Counter(preds)
classes = sorted(set(labels + preds))
true_vals = [true_count.get(i, 0) for i in classes]
pred_vals = [pred_count.get(i, 0) for i in classes]
x = [str(c) for c in classes]

plt.figure(figsize=(8, 5))
plt.bar(x, true_vals, label='Actual', alpha=0.6)
plt.bar(x, pred_vals, label='Predicted', alpha=0.6)
plt.xlabel("Class Index")
plt.ylabel("Count")
plt.title("üìä Actual vs Predicted Class Distribution")
plt.legend()
plt.show()


sampled = list(zip(X_test.tolist(), preds, y_test.tolist()))
random.shuffle(sampled)
print("\nüìã Sample Predictions:")
for seq, pred, true in sampled[:5]:
    true_label = class_mapping.get(true, str(true))
    pred_label = class_mapping.get(pred, str(pred))
    correct = "‚úÖ" if true == pred else "‚ùå"
    print(f"üß¨ Sequence: {seq[:50]}... | Predicted: {pred_label}, Actual: {true_label} {correct}")
```
